```python
def Iowa_stat_by_cat(category, num_col, func, other_categories = []):
    result = pd.DataFrame(iowamaster.groupby(category)[num_col].apply(func))

    
    for other in other_categories:
        temp = pd.DataFrame(iowamaster.groupby(category)[other].agg(lambda x:x.value_counts().index[0]))
        result = result.merge(temp, left_index=True, right_index=True)

    
    result = result.sort_values(by = num_col, ascending = False)
    
    return(result)
#All credit goes to Ben Shaver for the this wonderful bit of query code. Man's a genius at this stuff.
```

Seeing whether as Population of Age 21 or older increases, does the bottle price increase as a result.

```python
import statsmodels.api as sm

X = scaled_iowanum['Total 21 and Over Population']
y = scaled_iowanum['Sale (Dollars)']

model = sm.OLS(X, y).fit()
predictions = model.predict(X)

fig, ax = plt.subplots()
fig.set_size_inches(12, 12)

sns.regplot(predictions, y, ci = None)
plt.xlabel("Predicted Values from Model")
plt.ylabel("Actual Values")
plt.title('Effect of Age on Total Sales')
plt.show()

print("MSE:", model.mse_model)
model.summary()
```
